```python
import nltk
import string
import re
nltk.download('stopwords')
```

```python
def text_lowercase(text):
    return text.lower()
def remove_numbers(text):
    result = re.sub(r'\d+', '', text)
    return result
def remove_punctuation(text):
    translator = str.maketrans('', '', string.punctuation)
    return text.translate(translator)
def remove_whitespace(text):
    return  " ".join(text.split())
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk import pos_tag
nltk.download('averaged_perceptron_tagger')
```

```python
def remove_stopwords(text):
    stop_words = set(stopwords.words("english"))
    word_tokens = word_tokenize(text)
    filtered_text = [word for word in word_tokens if word not in stop_words]
    return filtered_text
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
# lemmatize string
def lemmatization(text, tags=['RB', 'JJ','NN','VBG']): # filter noun and adjective
       return [lemmatizer.lemmatize(token) for token in text if pos_tag([token])[0][1] in tags]
```

```python
def text_processing(inputs):  
  Processed_text=[]
  for sentence in inputs:
      x = text_lowercase(sentence)
      x = remove_numbers(x)
      x = remove_punctuation(x)
      x = remove_whitespace(x)
      x = remove_stopwords(x)
      x = lemmatization(x)
      Processed_text.append(x)
  return Processed_text  
```

```python
import gensim
from gensim import corpora
import pyLDAvis
```

```python
T=[]
for t in text:
    T.append([t])
```

```python
len(T)
```

## LDA 

```python
dictionary = corpora.Dictionary(T)
```

```python
doc_term_matrix = [dictionary.doc2bow(rev) for rev in T]
```

```python
# Creating the object for LDA model using gensim library
LDA = gensim.models.ldamodel.LdaModel

# Build LDA model
lda_model = LDA(corpus=doc_term_matrix, id2word=dictionary, num_topics=15, random_state=100,
                chunksize=1000, passes=50)
```

```python
lda_model.print_topics()
```

```python
# Visualize the topics
import pyLDAvis.gensim_models
pyLDAvis.enable_notebook()
#vis = pyLDAvis.gensim_models.prepare(lda_model, doc_term_matrix, dictionary)
topicData = pyLDAvis.gensim_models.prepare(lda_model, doc_term_matrix, dictionary, mds='mmds')   
topicData
```

```python
Topics = []
for i in range(15):
 Extract =[]
 for j in range(len(lda_model.get_topic_terms(i))):   
  cor = lda_model.get_topic_terms(i)[j][0]
  for Id,tex in dictionary.id2token.items():
        if cor==Id:
           Extract.append(tex)
 Topics.append(Extract)
    
```
